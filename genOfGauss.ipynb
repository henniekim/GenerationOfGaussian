```python
g1_lena = np.zeros((lena_w, lena_h))

for i in range(lena_h):
    for j in range(lena_w):
        output = lena[i,j] + int(np.random.normal(0, 50))

        # clipping
        if output >= 255 :
            output = 255
        if output <= 0 :
            output =0 

        g1_lena[i,j] = output

plt.imshow(g1_lena, cmap='gray')
```

```python
plt.imshow(lena, cmap='gray')
```

```python
# Know originals
diff = g1_lena - lena
plt.hist(diff.ravel(),256,[0,256])
plt.show()
```
