```python
g1_lena = np.zeros((lena_w, lena_h))

for i in range(lena_h):
    for j in range(lena_w):
        output = lena[i,j] + int(np.random.normal(0, 100))

        # clipping
        if output >= 255 :
            output = 255
        if output <= 0 :
            output =0 

        g1_lena[i,j] = output

plt.imshow(g1_lena[150:180,50:90], cmap='gray')
```

```python
# Know originals
diff = g1_lena - lena
n,b,p = plt.hist(g1_lena[150:180,50:90].ravel(),256,[0,256], normed=1, cumulative=1)
plt.show()
bin_ = np.where(n>=0.68)

print(np.where(n>=0.5))
print(np.where(n>=0.16))
print (bin_)

plt.hist(g1_lena[150:180,50:90].ravel(),256,[0,256], normed=1, cumulative=0)
plt.show()

```
